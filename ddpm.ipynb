```python
import numpy as np 

class_change = {}

for t in np.unique(df_metrics['transfer']):
    
    df_transfer = df_metrics.loc[(
        df_metrics['transfer'] == t
    )]

    class_change[t] = 0

    for i in range(len(df_transfer)):
        if df_transfer['gen_pred'].iloc[i] == df_transfer['target_pred'].iloc[i]:
            class_change[t] += 1

    class_change[t] = class_change[t]/len(df_transfer)
```

## Correlation with 0.2 threshold

```python
import numpy as np
def get_correlation(inim, outim):
    '''
    Compute the Pearson's correlation coefficient between original and reconstructed images.
    '''
    #orig, repro = utils.mask_using_original(inim, outim)
    
    data1 = inim.get_fdata().copy()
    data2 = outim.get_fdata().copy()
    
    # Vectorise input data
    data1 = np.reshape(data1, -1)
    data2 = np.reshape(data2, -1)

    in_mask_indices = np.logical_not(
        np.logical_or(
            np.logical_or(np.isnan(data1), np.absolute(data1) < 0.2),
            np.logical_or(np.isnan(data2), np.absolute(data2) < 0.2)))

    data1 = data1[in_mask_indices] 
    data2 = data2[in_mask_indices]
    
    corr_coeff = np.corrcoef(data1, data2)[0][1]
    
    return corr_coeff
```

```python
import nibabel as nib
from glob import glob

corr_st = {}
corr_sg = {}
corr_tg = {}

for w in [0, 1.0, 2.0]:
    corr_st[w] = []
    corr_tg[w] = []
    corr_sg[w] = []
    
    src_list = sorted(glob(f'/Volumes/egermani/star_ddpm/samples-2classes/src-image_*-dataset_rh_2class-jeanzay_ep140_w{w}-orig_0-target_1.nii.gz'))
    trg_list = sorted(glob(f'/Volumes/egermani/star_ddpm/samples-2classes/trg-image_*-dataset_rh_2class-jeanzay_ep140_w{w}-orig_0-target_1.nii.gz'))
    gen_list = sorted(glob(f'/Volumes/egermani/star_ddpm/samples-2classes/gen-image_*-dataset_rh_2class-jeanzay_ep140_w{w}-orig_0-target_1.nii.gz'))

    for i in range(len(gen_list)):
        src = nib.load(src_list[i])
        trg = nib.load(trg_list[i])
        gen = nib.load(gen_list[i])

        corr_st[w].append(get_correlation(src, trg))
        corr_sg[w].append(get_correlation(src, gen))
        corr_tg[w].append(get_correlation(gen, trg))
```
